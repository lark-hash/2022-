(1) 根据表 7，构建新老客户活跃程度的特征，并将结果存于“IsActiveStatus”
列。

(2) 根据表 8，构建不同金融资产客户活跃程度的特征，并将结果存于
“IsActiveAssetStage”列。

(3) 根据表 9，构建不同金融资产信用卡持有状态的特征，并将结果存于
“CrCardAssetStage”列。

```python
import pandas as pd
```

```python
df_long = pd.read_excel('../result/result3.xlsx')
```

```python
df_long.head(3)
```

```python
def IsActiveStatus(g,e):
    if ((g=='新客户') & (e==0)):
        return 0
    if (g=='稳定客户') & (e==0):
        return 1
    if (g=='老客户') & (e==0):
        return 2
    if (g=='新客户') & (e==1):
        return 3
    if (g=='稳定客户') & (e==1):
        return 4
    if (g=='老客户') & (e==1):
        return 5
```

```python
df_long['IsActiveStatus'] = df_long.apply(lambda x: IsActiveStatus(x['Status'], x['IsActiveMember']), axis=1)
```

```python
df_long.head(3)
```

```python

def IsActiveAssetStage(g,e):
    if ((g=='低资产') & (e==0)):
        return 0
    if (g=='中下资产') & (e==0):
        return 1
    if (g=='中上资产') & (e==0):
        return 2
    if (g=='高资产') & (e==0):
        return 3
    if (g=='低资产') & (e==1):
        return 6
    if (g=='中下资产') & (e==1):
        return 7
    if (g=='中上资产') & (e==1):
        return 8
    if (g=='高资产') & (e==1):
        return 9
```

```python
df_long['IsActiveAssetStage'] = df_long.apply(lambda x: IsActiveAssetStage(x['AssetStage'], x['IsActiveMember']), axis=1)
```

```python
df_long.head(3)
```

```python
def CrCardAssetStage(g,e):
    if ((g=='低资产') & (e==0)):
        return 0
    if (g=='中下资产') & (e==0):
        return 2
    if (g=='中上资产') & (e==0):
        return 5
    if (g=='高资产') & (e==0):
        return 5
    if (g=='低资产') & (e==1):
        return 6
    if (g=='中下资产') & (e==1):
        return 7
    if (g=='中上资产') & (e==1):
        return 9
    if (g=='高资产') & (e==1):
        return 9
```

```python
df_long['CrCardAssetStage'] = df_long.apply(lambda x: CrCardAssetStage(x['AssetStage'], x['HasCrCard']), axis=1)
```
