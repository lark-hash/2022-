## 任务1.1(1)

探索短期数据各指标数据的缺失值和“user_id”列重复值，删除缺失值、
重复值所在行数据。请在报告中给出处理过程及必要结果，完整的结果保存到文
件“result1_1.xlsx”中。


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_short = pd.read_csv('../short-customer-data.csv')
```

```python
df_short.head()
```

```python
# 找出含有缺失值的行
df_short[df_short.isna().any(axis=1)]
```

```python
# 删除含有缺失值的行
df_short.dropna(axis=0, how='any', inplace=True)
```

```python
df_short
```

```python
# 获取重复项索引
index_duplicated = df_short.loc[df_short['user_id'].duplicated()].index
```

```python
# 删除user_id列重复的行
df_short.drop(index=index_duplicated, inplace=True)
```

```python
# 任务1.1，保存到文件result1_1.xlsx中
df_short.to_excel('../result/result1_1.xlsx', index=False)
```

## 任务1.1(2)

长期数据中的客户年龄“Age”列存在数值为-1、0 和“-”的异常值，删
除存在该情况的行数据；“Age”列存在空格和“岁”等异常字符，删除这些异常
字符但须保留年龄数值，将处理后的数值存于“Age”列。请在报告中给出处理
过程及必要结果，完整的结果保存到文件“result1_2.xlsx”中。

```python
import pandas as pd
```

```python
df_long = pd.read_csv('../long-customer-train.csv')
```

```python
df_long
```

```python
df_long_error = df_long.loc[(df_long['Age'] == '-1') | (df_long['Age'] == '0') | (df_long['Age'] == '-') | (df_long['Age'] == '1')].index
```

```python
df_long.drop(index=df_long_error, inplace=True)
```

```python
df_long
```

```python
df_long['Age'] = df_long['Age'].str.extract('(\d+)', expand=False)
```

```python
df_long['Age'] = pd.to_numeric(df_long['Age'])
```

```python
df_long['Age'].dtype
```

```python
df_long[df_long['Age']=="0"]
```

```python
df_long.to_excel('../result/result1_2.xlsx', index=False)
```

## 任务1.2

任务 1.2 对短期数据中的字符型数据进行特征编码，如将信用违约情况
{‘否’,‘是’}编码为{0,1}。请在报告中给出处理思路、过程及必要结果，完整的结
果保存到文件“result1_3.xlsx”中。

```python
import pandas as pd
```

```python
df_short = pd.read_excel('../result/result1_1.xlsx')
```

```python
df_short.head()
```

```python
df_short['job'].value_counts()
```

```python
df_short['job'] = df_short['job'].replace({'unemployed' : 0, 'technician' : 1, 'student' : 2, 'services' : 3, 'self-employed' : 4,
                                           'retired' : 5, 'management' : 6, 'housemaid' : 7, 'entrepreneur' : 8, 'blue-collar' : 9,
                                           'admin.' : 10})
```

```python
df_short['marital'] = df_short['marital'].replace({'single' : 0, 'married' : 1, 'divorced' : 2})
```

```python
df_short['education'] = df_short['education'].replace({'illiterate':0, 'high school':1, 
                                                       'junior college':2, 'undergraduate':3, 'postgraduate':4})
```

```python
df_short['default'] = df_short['default'].replace({'no':0, 'yes':1})
df_short['housing'] = df_short['housing'].replace({'no':0, 'yes':1})
df_short['loan'] = df_short['loan'].replace({'no':0, 'yes':1})
df_short['contact'] = df_short['contact'].replace({'cellular':0, 'telephone':1})
```

```python
df_short['month'].value_counts()
```

```python
df_short['month'] = df_short['month'].replace({'mar':3, 'apr':4, 'may':5, 'jun':6, 'jul':7, 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec':12})
```

```python
df_short['day_of_week'].value_counts()
```

```python
df_short['day_of_week'] = df_short['day_of_week'].replace({'mon':1, 'thu':2, 'wed':3, 'tue':4, 'fri':5})
```

```python
df_short['poutcome'].value_counts()
```

```python
df_short['poutcome'] = df_short['poutcome'].replace({'failure':0, 'success':1, 'nonexistent':2})
```

```python
df_short['y'].value_counts()
```

```python
df_short['y'] = df_short['y'].replace({'no':0, 'yes':1})
```

```python
df_short
```

```python
df_short.to_excel('../result/result1_3.xlsx', index=False)
```
